## How let's make a function

$$
{\rm Free\_Fall\_H}(t) = \frac{1}{2} g t^2
$$

```python
def Free_Fall_H(t):
    g = 9.80665 # m/s
    
    H = 0.5 * g * t**2
    return H
```
